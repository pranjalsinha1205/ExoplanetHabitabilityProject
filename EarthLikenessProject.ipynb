# Rules for null values removal

***1) Null values < 10:*** .dropna()

***2) Null values in range (10, 50):*** .fillna()

***3) Null values >= 50:*** remove column

```python
try:
    df.drop(columns=['st_spectype'], inplace=True)
    df['pl_orbeccen'].fillna(df["pl_orbeccen"].median(), inplace=True)
    df['pl_eqt'].fillna(df["pl_eqt"].median(), inplace=True)
    df['pl_insol'].fillna(df["pl_insol"].median(), inplace=True)
    df.dropna(inplace=True)
    print(f"Null values taken care of properly, shape of the dataset: {df.shape}")
except Exception as e:
    print(e)
```

```python
def cap_outliers(df, features):
    df_capped = df.copy()
    for col in features:
        Q1 = df_capped[col].quantile(0.25)
        Q3 = df_capped[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_capped[col] = df_capped[col].clip(lower=lower_bound, upper=upper_bound)
    return df_capped
```

```python
features_to_clean = [
    'pl_orbsmax', 'pl_rade', 'pl_bmasse', 'pl_orbeccen',
    'pl_insol', 'pl_eqt', 'st_teff', 'st_rad', 'st_mass', 'sy_dist'
]

df_clean = cap_outliers(df, features_to_clean)
```

```python
features = ['pl_orbper', 'pl_orbsmax', 'pl_rade', 'pl_bmasse', 'pl_orbeccen', 'pl_insol',
            'pl_eqt', 'st_teff', 'st_rad', 'st_mass']
df = df[features]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df)
```

***Earth Reference***

```python
earth = {
    'pl_orbper': 365.25,
    'pl_orbsmax': 1.0,
    'pl_rade': 1.0,
    'pl_bmasse': 1.0,
    'pl_orbeccen': 0.0167,
    'pl_insol': 1.0,
    'pl_eqt': 288,
    'st_teff': 5778,
    'st_rad': 1.0,
    'st_mass': 1.0
}
```

Convert it into a vector

```python
earth_vector_raw = np.array([earth[feat] for feat in features]).reshape(1, -1)
earth_vector_scaled = scaler.transform(earth_vector_raw)[0]
```

```python
from scipy.spatial.distance import euclidean

similarity_scores = []
for row in scaled_features:
    dist = euclidean(row, earth_vector_scaled)
    similarity = 1 - dist  # Closer to 1 means more similar to Earth
    similarity_scores.append(similarity)

df['earth_similarity'] = similarity_scores
```

# 2nd part of the project

```python
from sklearn.cluster import KMeans

inertias = []
K_range = range(1, 10)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertias.append(kmeans.inertia_)

# Plot the elbow
plt.plot(K_range, inertias, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

df['cluster'] = clusters
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced_features = pca.fit_transform(scaled_features)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(reduced_features[:, 0], reduced_features[:, 1],
                      c=df['cluster'], cmap='viridis', alpha=0.6)
plt.colorbar(scatter)
plt.title('KMeans Clustering of Exoplanets (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```

```python
# Assume df_clean is your cleaned DataFrame and scaler is already fitted
earth_values = df_clean[features].mean().values  # or use Earth-like values manually
earth_scaled = scaler.transform([earth_values])[0]

variation_pct = np.linspace(0.5, 1.5, 50)
sensitivity = {}

for i, feature in enumerate(features):
    scores = []
    for pct in variation_pct:
        # Copy actual Earth values and modify one
        modified_vals = earth_values.copy()
        modified_vals[i] *= pct

        # Scale using fitted scaler
        modified_scaled = scaler.transform([modified_vals])[0]

        # Calculate similarity
        dist = euclidean(modified_scaled, earth_scaled)
        score = 1 - dist
        scores.append(score)
    
    sensitivity[feature] = scores
```

```python
# Plot sensitivity curves
plt.figure(figsize=(12, 6))
for feature in features:
    plt.plot(variation_range, sensitivity[feature], label=feature)

plt.title("Feature Sensitivity Analysis on Earth Similarity Score")
plt.xlabel("Relative Change in Feature (Earth = 1.0)")
plt.ylabel("Similarity Score (1 - Distance)")
plt.legend(loc='best', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
features = ['pl_orbper', 'pl_orbsmax', 'pl_rade', 'pl_bmasse', 'pl_orbeccen',
            'pl_insol', 'pl_eqt', 'st_teff', 'st_rad', 'st_mass']

X = df[features].copy()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
kmeans.fit(X_scaled)
```

```python
earth_like = pd.DataFrame([{
    'pl_orbper': 365.25,
    'pl_orbsmax': 1.0,
    'pl_rade': 1.0,
    'pl_bmasse': 1.0,
    'pl_orbeccen': 0.0167,
    'pl_insol': 1.0,
    'pl_eqt': 288,
    'st_teff': 5778,
    'st_rad': 1.0,
    'st_mass': 1.0
}])

earth_scaled = scaler.transform(earth_like)
```

```python
earth_cluster = kmeans.predict(earth_scaled)[0]
earth_cluster_center = kmeans.cluster_centers_[earth_cluster]
```

```python
all_distances = np.linalg.norm(X_scaled - earth_cluster_center, axis=1)
max_dist = np.max(all_distances)
```

```python
import joblib

try:
    joblib.dump({
        'kmeans': kmeans,
        'scaler': scaler,
        'earth_cluster_center': earth_cluster_center,
        'features': features,
        'max_dist': max_dist
    }, 'earth_similarity_model.joblib')

    print(f"Model saved as 'earth_similarity_model.joblib'. Earth-like cluster: {earth_cluster}")
except Exception as e:
    print(e)
```
