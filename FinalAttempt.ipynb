***Key Features for Determining Habitability:***

1) Insolation Flux (pl_insol): The amount of stellar radiation received by the planet. It's usually compared to Earth’s value (1 Earth Flux = Earth’s insolation). Habitable range: 0.75 - 1.25 Earth Flux (for liquid water to potentially exist).

2) Equilibrium Temperature (pl_eqt): Temperature the planet would have if it were in thermal equilibrium (no internal heat sources).

3) Habitable range: 273 K (0°C) to 373 K (100°C) for water to exist in liquid form.

4) Planetary Size (pl_rade): Larger planets may have a thick atmosphere, making them less likely to be habitable, while smaller planets may lack the necessary atmosphere for life. Earth-like planets have a radius close to 1 Earth Radius.

5) Orbital Period (pl_orbper) and Semi-Major Axis (pl_orbsmax): These help determine the distance from the star and thus influence the climate. A longer orbital period could indicate the planet is farther from its star.

We have already dropped pl_orbper because it was perfectly correlated with orbsmax

```python
def is_habitable(row):
    """
    Determines whether a planet is habitable based on its features.
    """
    # Habitable zone criteria
    # Orbital period is already dropped, so we use orbit semi-major axis (pl_orbsmax) and other factors
    orbit_semi_major_axis = row['pl_orbsmax']  # In AU
    radius = row['pl_rade']  # In Earth radii
    mass = row['pl_bmasse']  # In Earth masses
    eccentricity = row['pl_orbeccen']  # Eccentricity
    insolation_flux = row['pl_insol']  # In Earth flux
    equilibrium_temp = row['pl_eqt']  # In Kelvin
    stellar_temp = row['st_teff']  # Stellar effective temperature in Kelvin
    stellar_radius = row['st_rad']  # Stellar radius in Solar radii
    stellar_mass = row['st_mass']  # Stellar mass in Solar masses
    distance = row['sy_dist']  # Distance in parsecs

    # Define habitability conditions based on common understanding:
    
    # Condition 1: Planet should be within the habitable zone (defined by its orbit and insolation)
    # A general criterion for the habitable zone:
    if not (0.5 < orbit_semi_major_axis < 2.0):  # Orbital range (AU)
        return 0
    
    # Condition 2: The planet should have a radius similar to Earth
    if radius < 0.5 or radius > 2.0:  # In Earth radii
        return 0
    
    # Condition 3: Planet mass should be in Earth-like range
    if mass < 0.5 or mass > 10.0:  # In Earth masses
        return 0
    
    # Condition 4: Orbital eccentricity should be low for stable climates
    if eccentricity > 0.4:
        return 0
    
    # Condition 5: The planet should have an equilibrium temperature that is Earth-like (0°C to 50°C)
    if equilibrium_temp < 223 or equilibrium_temp > 343:  # K (0°C to 50°C)
        return 0
    
    # Condition 6: Insolation flux should be similar to Earth's (between 0.5 to 2.0 Earth flux)
    if insolation_flux < 0.5 or insolation_flux > 2.0:
        return 0
    
    # If all conditions are satisfied, the planet is considered habitable
    return 1
```

```python
df['habitable'] = (
    (df['pl_orbsmax'].between(0.38, 2.0)) &
    (df['pl_rade'].between(0.5, 2.5)) &
    (df['pl_bmasse'].between(0.1, 10)) &
    (df['pl_orbeccen'] < 0.5) &
    (df['pl_insol'].between(0.35, 1.75)) &
    (df['pl_eqt'].between(180, 310)) &
    (df['st_teff'].between(2600, 7200)) &
    (df['st_rad'].between(0.1, 2.0)) &
    (df['st_mass'].between(0.1, 2.0)) &
    (df['sy_dist'] < 100)
)
```

```python
df['habitable'].value_counts()
```

```python
features = [
    'pl_orbsmax', 'pl_rade', 'pl_bmasse', 'pl_orbeccen', 'pl_insol',
    'pl_eqt', 'st_teff', 'st_rad', 'st_mass', 'sy_dist'
]

# Select the relevant features and target
X = df[features]
y = df['habitable']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='auto', random_state=42, k_neighbors=1)
X_res, y_res = smote.fit_resample(X_scaled, y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create a heatmap to visualize the confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Habitable', 'Habitable'], yticklabels=['Non-Habitable', 'Habitable'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
def cap_outliers(df, features):
    df_capped = df.copy()
    for col in features:
        Q1 = df_capped[col].quantile(0.25)
        Q3 = df_capped[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_capped[col] = df_capped[col].clip(lower=lower_bound, upper=upper_bound)
    return df_capped
```
