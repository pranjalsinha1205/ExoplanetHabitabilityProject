```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df = pd.read_excel("planets_data.xlsx")
df = pd.read_csv("ExoplanetHabitabilityData.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
null_values = (df.isnull().sum()/len(df))*100
null_values
```

```python
df1 = df.copy()
```

***Dealing with null values***

```python
df1 = df1.loc[:, null_values <= 50]
fillna_columns = null_values[(null_values > 10) & (null_values <= 50)].index
df1[fillna_columns] = df1[fillna_columns].fillna(df1[fillna_columns].median()) 
df1.dropna(inplace=True)
```

```python
df.info()
```

```python
df1.info()
```

```python
df1.shape
```

```python
df1.drop_duplicates(inplace=True)
```

```python
df1.shape
```

```python
df1.describe(include="all")
```

```python
from scipy import stats
```

```python
numeric_df = df1.select_dtypes(include=['int64', 'float64'])
```

```python
z_scores = np.abs(stats.zscore(numeric_df))
z_scores
```

```python
df_no_outliers = df1[(z_scores < 3).all(axis=1)]
print("\nDataFrame after removing outliers:")
df_no_outliers
```
